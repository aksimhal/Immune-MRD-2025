```python
import ot
import scipy
import umap
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Example code to generate random numbers from two 2D Gaussian distributions
# and visualize them using scatter plots and kernel density estimates.
# Define the parameters for the first distribution
mean1 = [0, 0]
covariance1 = [[1, 0], [0, 1]]

# Define the parameters for the second distribution (offset by 2 in both x and y)
mean2 = [2, 2]
covariance2 = [[1, 0], [0, 1]]

# Generate random samples from the first distribution
num_samples = 500
distribution1 = np.random.multivariate_normal(mean1, covariance1, num_samples)

# Generate random samples from the second distribution
distribution2 = np.random.multivariate_normal(mean2, covariance2, num_samples)

# Separate the x and y coordinates for plotting
x1, y1 = distribution1[:, 0], distribution1[:, 1]
x2, y2 = distribution2[:, 0], distribution2[:, 1]

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x1, y1, label='Distribution 1', alpha=0.6)
plt.scatter(x2, y2, label='Distribution 2', alpha=0.6)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two 2D Distributions Offset by 2')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Show the plot
plt.show()
```

```python
# Exammple code to compute the Wasserstein distance between two 2D distributions
ot.sliced_wasserstein_distance(distribution1, distribution2)
```
